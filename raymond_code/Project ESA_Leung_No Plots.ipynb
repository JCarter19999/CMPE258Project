```python
import pandas as pdyujk,
import numpy as np
from sklearn.preprocessing import StandardScaler
import umap
import umap.umap_ as umap # Explicitly import umap.umap_ as umap
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

print("All visualization libraries imported successfully.")
```

```python
#Define  visualize_clusters includeing data extraction, scaling, UMAP and t-SNE dimensionality reduction, and interactive scatter plot generation


def visualize_clusters(df_sample, feature_columns, title_prefix):
    """
    Generates and displays UMAP and t-SNE interactive scatter plots for sampled data.

    Args:
        df_sample (pd.DataFrame): DataFrame containing sampled data and 'is_anomaly' labels.
        feature_columns (list): List of column names to use as features.
        title_prefix (str): Prefix for the plot titles.
    """
    print(f"Generating visualizations for: {title_prefix} with {len(feature_columns)} features.")

    # 1. Extract features and 'is_anomaly' column
    X = df_sample[feature_columns]
    y = df_sample['is_anomaly']

    # 2. Scale the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # 3. Apply UMAP to reduce dimensions to 2
    reducer_umap = umap.UMAP(n_components=2, random_state=42)
    umap_results = reducer_umap.fit_transform(X_scaled)

    # 4. Apply t-SNE to reduce dimensions to 2
    reducer_tsne = TSNE(n_components=2, random_state=42, perplexity=30)
    tsne_results = reducer_tsne.fit_transform(X_scaled)

    # 7. Create a temporary DataFrame for UMAP results
    umap_df = pd.DataFrame(umap_results, columns=['UMAP-1', 'UMAP-2'])
    umap_df['is_anomaly'] = y.astype(str).reset_index(drop=True) # Convert to string for discrete colors

    # 8. Generate interactive scatter plot for UMAP
    fig_umap = px.scatter(
        umap_df,
        x='UMAP-1',
        y='UMAP-2',
        color='is_anomaly',
        title=f'{title_prefix} UMAP Visualization ({len(feature_columns)} features)',
        hover_data={'is_anomaly': True}
    )
    fig_umap.show()

    # 9. Create a temporary DataFrame for t-SNE results
    tsne_df = pd.DataFrame(tsne_results, columns=['TSNE-1', 'TSNE-2'])
    tsne_df['is_anomaly'] = y.astype(str).reset_index(drop=True) # Convert to string for discrete colors

    # 10. Generate interactive scatter plot for t-SNE
    fig_tsne = px.scatter(
        tsne_df,
        x='TSNE-1',
        y='TSNE-2',
        color='is_anomaly',
        title=f'{title_prefix} t-SNE Visualization ({len(feature_columns)} features)',
        hover_data={'is_anomaly': True}
    )
    fig_tsne.show()

print("Visualization function 'visualize_clusters' defined successfully.")
```

# Model and Training

Use list of target channel names from target_channels_df to be used as features for the LSTM model. These will be the input columns for model.


```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.utils import Sequence

print("Libraries imported successfully: numpy, MinMaxScaler, Sequence.")
```

```python
# Define AnomalyDataGenerator class
# Will handle data selection, normalization, sequence reshaping, and label extraction in batches for both training and validation sets


class AnomalyDataGenerator(Sequence):
    def __init__(self, df, target_channels, time_steps, batch_size, validation_split_ratio, is_validation_set=False, scaler=None):
        self.df = df
        self.target_channels = target_channels
        self.time_steps = time_steps
        self.batch_size = batch_size
        self.validation_split_ratio = validation_split_ratio
        self.is_validation_set = is_validation_set

        # Determine split index
        split_index = int(len(self.df) * (1 - self.validation_split_ratio))

        if not self.is_validation_set:
            self.data_df = self.df.iloc[:split_index].copy()
        else:
            self.data_df = self.df.iloc[split_index:].copy()

        if scaler is None:
            self.scaler = MinMaxScaler()
            self.scaler.fit(self.data_df[self.target_channels])
            print("MinMaxScaler fitted on training data.")
        else:
            self.scaler = scaler
            print("Using provided MinMaxScaler.")

        self.on_epoch_end()

    def __len__(self):
        # Number of batches in the dataset
        return (len(self.data_df) - self.time_steps) // self.batch_size

    def __getitem__(self, index):
        start_index = index * self.batch_size
        end_index = start_index + self.batch_size

        # Ensure we don't go out of bounds considering time_steps
        if end_index + self.time_steps > len(self.data_df):
            end_index = len(self.data_df) - self.time_steps
            start_index = end_index - self.batch_size
            if start_index < 0: # Handle case where remaining data is less than batch_size
                start_index = 0
                end_index = len(self.data_df) - self.time_steps
                if end_index < 0: # No full sequences possible
                    raise IndexError("Not enough data to form sequences for this batch.")

        batch_data = self.data_df.iloc[start_index : end_index + self.time_steps]

        # Select target channels and normalize
        features = self.scaler.transform(batch_data[self.target_channels])

        # Reshape to sequences
        X = np.array([features[i : i + self.time_steps] for i in range(self.batch_size)])

        # Extract labels (is_anomaly at the end of each sequence)
        y = batch_data['is_anomaly'].iloc[self.time_steps - 1 : self.time_steps - 1 + self.batch_size].values

        return X, y

    def on_epoch_end(self):
        # Optional: shuffle data if needed. For time series, usually we don't shuffle
        # but for demonstration we can show where it would go.
        pass # No shuffling for time series to maintain order

print("AnomalyDataGenerator class defined successfully.")
```

```python
# With the AnomalyDataGenerator class defined, instantiate two instances of it:
# one for the training dataset and one for the validation dataset.
# Set time_steps to 60, batch_size to 512, and validation_split_ratio to 0.2.
# Scaler will be fitted on the training data and then passed to the validation generator to ensure consistent scaling.


time_steps = 60
batch_size = 512
validation_split_ratio = 0.2

# Create the training data generator and fit the scaler
train_generator = AnomalyDataGenerator(
    train_df,
    target_channels,
    time_steps,
    batch_size,
    validation_split_ratio,
    is_validation_set=False
)

# Create the validation data generator using the scaler fitted on the training data
validation_generator = AnomalyDataGenerator(
    train_df,
    target_channels,
    time_steps,
    batch_size,
    validation_split_ratio,
    is_validation_set=True,
    scaler=train_generator.scaler
)

print(f"Train generator created with {len(train_generator)} batches.")
print(f"Validation generator created with {len(validation_generator)} batches.")
print(f"Validation split ratio: {validation_split_ratio}")
```

```python
# TestDataGenerator` class, to prepare the test data for the LSTM model.
# This class will handle selecting target channels, normalizing data using the scaler fitted on the training data,

import numpy as np
from tensorflow.keras.utils import Sequence

class TestDataGenerator(Sequence):
    def __init__(self, df, target_channels, time_steps, batch_size, scaler):
        self.df = df
        self.target_channels = target_channels
        self.time_steps = time_steps
        self.batch_size = batch_size
        self.scaler = scaler

        # No validation split for test data
        self.data_df = self.df.copy()

        print("TestDataGenerator initialized.")

    def __len__(self):
        # Number of batches in the dataset. Ensure full sequences can be formed.
        return max(0, (len(self.data_df) - self.time_steps + 1) // self.batch_size)

    def __getitem__(self, index):
        start_index = index * self.batch_size
        end_index = start_index + self.batch_size

        # Ensure we don't go out of bounds considering time_steps
        # We need data from start_index to end_index + time_steps - 1 to form all sequences in the batch
        # The range is [start_index, end_index + self.time_steps - 1]

        # Adjust end_index for slicing to make sure we have enough data for the last sequence of the batch
        # The last element in the batch of sequences would be at `end_index + time_steps - 1`
        actual_end_slice = min(len(self.data_df), end_index + self.time_steps -1)
        actual_start_slice = start_index

        # If the actual_end_slice does not allow for a full batch of sequences,
        # adjust the start_index backwards to get a full batch if possible, or take what's left
        # This logic is a bit tricky with `Sequence` for the very last batch.
        # A simpler approach for test data might be to generate all sequences and then batch them.
        # For now, let's ensure we grab enough for `batch_size` sequences.

        # Correct logic for extracting data for X sequences from a contiguous block:
        # We need 'batch_size' sequences. Each sequence is 'time_steps' long.
        # The first sequence starts at 'start_index' and ends at 'start_index + time_steps - 1'.
        # The last sequence starts at 'start_index + batch_size - 1' and ends at 'start_index + batch_size - 1 + time_steps - 1'.
        # So, we need data from `start_index` to `start_index + batch_size + time_steps - 2` (inclusive).

        data_to_process = self.data_df.iloc[start_index : start_index + self.batch_size + self.time_steps - 1]

        # Select target channels and normalize
        features_raw = data_to_process[self.target_channels].values
        features_scaled = self.scaler.transform(features_raw)

        # Reshape to sequences
        X = np.array([features_scaled[i : i + self.time_steps] for i in range(self.batch_size)])

        return X

    def on_epoch_end(self):
        pass # No shuffling for time series to maintain order

# Instantiate TestDataGenerator
test_generator = TestDataGenerator(
    test_df,
    target_channels,
    time_steps,
    batch_size,
    scaler=train_generator.scaler
)

print(f"Test generator created with {len(test_generator)} batches.")
```

```python
#Calculate class weights based on the is_anomaly distribution in train_df.
# Will be used during model training to address the highly imbalanced dataset, giving more importance to the minority class (anomalies).

from sklearn.utils.class_weight import compute_class_weight
import numpy as np

print("Libraries imported successfully: compute_class_weight, numpy.")
```

```python
#Calculate weight classes

labels = train_df['is_anomaly'].values
classes = np.unique(labels)

class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=labels
)

class_weights = dict(zip(classes, class_weights_array))

print(f"Calculated class weights: {class_weights}")
```

## Define LSTM Model Architecture

Handle sequential input data with the number of features and time_steps as determined in previous steps. Include appropriate layers and an output layer for binary classification.


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input

input_shape = (time_steps, len(target_channels))

model = Sequential([
    Input(shape=input_shape), # Explicitly define input layer
    LSTM(64, activation='relu', return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

model.summary()

print("LSTM model architecture defined successfully, warning addressed.")
```

## Compile and Train LSTM Model

Compile the LSTM model with an appropriate optimizer, a loss function suitable for binary classification, and relevant metrics. Train the model using the prepared training data generator, validation data, batch_size (which is 512), epochs (which is 20), and the calculated class weights to mitigate the class imbalance.


```python
from tensorflow.keras.metrics import Precision, Recall, AUC
from tensorflow.keras.callbacks import EarlyStopping

epochs = 20

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', Precision(), Recall(), AUC()]
)

print("Model compiled successfully.")

# Define EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

print("EarlyStopping callback defined.")

# Train the model
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator,
    class_weight=class_weights,
    callbacks=[early_stopping]
)

print("Model training complete.")
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2

print("Adam optimizer and l2 regularizer imported successfully.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.metrics import Precision, Recall, AUC
from tensorflow.keras.callbacks import EarlyStopping

# Redefine the LSTM model with updated architecture
model = Sequential([
    Input(shape=input_shape),
    LSTM(64, activation='tanh', kernel_regularizer=l2(0.001), return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

# Compile the revised model with a lower learning rate and clipvalue
model.compile(
    optimizer=Adam(learning_rate=0.0001, clipvalue=1.0),
    loss='binary_crossentropy',
    metrics=['accuracy', Precision(), Recall(), AUC()]
)

model.summary()

print("LSTM model architecture redefined and compiled successfully with Adam optimizer.")

# Define EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Train the model again
history_revised = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator,
    class_weight=class_weights,
    callbacks=[early_stopping]
)

print("Revised model training complete.")
```

## Parse Epoch Logs

### Subtask:
Extract the training and validation metrics (accuracy, AUC, loss, precision, and recall) for each epoch from the provided text logs. This data will be structured for plotting.


```python
import re
import pandas as pd

print("Libraries 're' and 'pandas' imported successfully.")
```

```python
import re
import pandas as pd

def parse_training_logs(log_text, model_type):
    epochs_data = []

    # Regex to clean ANSI escape codes
    ansi_escape = re.compile(r'\x1b\[[0-9;]*m')

    # Regex to extract epoch number (e.g., "Epoch 1/20")
    epoch_number_pattern = re.compile(r"^Epoch (\d+)/\d+")

    # Regex to capture all metrics from a line that contains 'val_accuracy'
    # Using named groups for clarity and robustness against order changes if present in the same line
    # Handles 'metric' and 'metric_1' variations and scientific notation (e.g., 0.0000e+00)
    metric_line_pattern = re.compile(
        r".*?accuracy(?:_\d+)?: (?P<train_accuracy>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?auc(?:_\d+)?: (?P<train_auc>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?loss: (?P<train_loss>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?precision(?:_\d+)?: (?P<train_precision>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?recall(?:_\d+)?: (?P<train_recall>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?val_accuracy(?:_\d+)?: (?P<val_accuracy>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?val_auc(?:_\d+)?: (?P<val_auc>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?val_loss: (?P<val_loss>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?val_precision(?:_\d+)?: (?P<val_precision>\d+\.?\d*(?:e[+\-]?\d+)?)"
        r".*?val_recall(?:_\d+)?: (?P<val_recall>\d+\.?\d*(?:e[+\-]?\d+)?)"
    )

    current_epoch = None
    for line in log_text.split('\n'):
        cleaned_line = ansi_escape.sub('', line).strip()

        epoch_match = epoch_number_pattern.match(cleaned_line)
        if epoch_match:
            current_epoch = int(epoch_match.group(1))
            continue

        # Look for the line containing validation metrics, which usually indicates the end of an epoch summary.
        # This line will also contain all training metrics for that epoch.
        if "val_accuracy" in cleaned_line and current_epoch is not None:
            metrics_match = metric_line_pattern.search(cleaned_line)
            if metrics_match:
                metrics = metrics_match.groupdict()
                epochs_data.append({
                    'epoch': current_epoch,
                    'model_type': model_type,
                    'train_accuracy': float(metrics.get('train_accuracy', 0.0)),
                    'train_auc': float(metrics.get('train_auc', 0.0)),
                    'train_loss': float(metrics.get('train_loss', 0.0)),
                    'train_precision': float(metrics.get('train_precision', 0.0)),
                    'train_recall': float(metrics.get('train_recall', 0.0)),
                    'val_accuracy': float(metrics.get('val_accuracy', 0.0)),
                    'val_auc': float(metrics.get('val_auc', 0.0)),
                    'val_loss': float(metrics.get('val_loss', 0.0)),
                    'val_precision': float(metrics.get('val_precision', 0.0)),
                    'val_recall': float(metrics.get('val_recall', 0.0))
                })
                current_epoch = None # Reset for the next epoch

    return pd.DataFrame(epochs_data)

# Retrieve raw standard output text from previous cells directly from the notebook state.
# The content is provided in the problem description for cell f28765a4 and 62021636.
initial_log_text = """Model compiled successfully.\nEarlyStopping callback defined.\nEpoch 1/20\n\n\n\x1b[1m23011/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m━\x1b[0m \x1b[1m0s\x1b[0m 10ms/step - accuracy: 0.5503 - auc: 0.5255 - loss: 0.7468 - precision: 0.1105 - recall: 0.4741\n\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m269s\x1b[0m 11ms/step - accuracy: 0.5503 - auc: 0.5255 - loss: 0.7468 - precision: 0.1105 - recall: 0.4741 - val_accuracy: 0.7856 - val_auc: 0.3076 - val_loss: 0.6805 - val_precision: 0.0093 - val_recall: 0.0099\nEpoch 2/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.6421 - auc: 0.5575 - loss: 181393.5156 - precision: 0.1285 - recall: 0.4114 - val_accuracy: 0.8947 - val_auc: 0.5000 - val_loss: 0.6825 - val_precision: 0.0000e+00 - val_recall: 0.0000e+00\nEpoch 3/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m258s\x1b[0m 11ms/step - accuracy: 0.3481 - auc: 0.4879 - loss: 6949.3530 - precision: 0.1057 - recall: 0.6756 - val_accuracy: 0.8947 - val_auc: 0.5000 - val_loss: 0.6642 - val_precision: 0.0000e+00 - val_recall: 0.0000e+00\nEpoch 4/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m261s\x1b[0m 11ms/step - accuracy: 0.7124 - auc: 0.4866 - loss: 0.8964 - precision: 0.1080 - recall: 0.2393 - val_accuracy: 0.1053 - val_auc: 0.5000 - val_loss: 0.6979 - val_precision: 0.1053 - val_recall: 1.0000\nEpoch 5/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m259s\x1b[0m 11ms/step - accuracy: 0.4266 - auc: 0.4915 - loss: 0.7428 - precision: 0.1028 - recall: 0.5811 - val_accuracy: 0.1884 - val_auc: 0.3132 - val_loss: 0.6960 - val_precision: 0.0711 - val_recall: 0.5561\nEpoch 6/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.3650 - auc: 0.5128 - loss: 1336.1749 - precision: 0.1094 - recall: 0.7093 - val_accuracy: 0.1053 - val_auc: 0.5000 - val_loss: 0.6937 - val_precision: 0.1053 - val_recall: 1.0000\nEpoch 7/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.6747 - auc: 0.5195 - loss: 188.0962 - precision: 0.1137 - recall: 0.3179 - val_accuracy: 0.8123 - val_auc: 0.7386 - val_loss: 0.6588 - val_precision: 0.2675 - val_recall: 0.4503\nEpoch 8/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m259s\x1b[0m 11ms/step - accuracy: 0.6329 - auc: 0.5045 - loss: 155184992.0000 - precision: 0.1092 - recall: 0.3506 - val_accuracy: 0.8947 - val_auc: 0.3182 - val_loss: 0.5837 - val_precision: 0.0000e+00 - val_recall: 0.0000e+00\nEpoch 9/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.4736 - auc: 0.5358 - loss: 0.7114 - precision: 0.1099 - recall: 0.5969 - val_accuracy: 0.8389 - val_auc: 0.5847 - val_loss: 0.8409 - val_precision: 0.2896 - val_recall: 0.3649\nEpoch 10/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.5658 - auc: 0.5685 - loss: 0.7668 - precision: 0.1239 - recall: 0.5168 - val_accuracy: 0.6892 - val_auc: 0.5968 - val_loss: 1.2970 - val_precision: 0.1571 - val_recall: 0.4469\nEpoch 11/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.3738 - auc: 0.5419 - loss: 0.7054 - precision: 0.1143 - recall: 0.7260 - val_accuracy: 0.5376 - val_auc: 0.4600 - val_loss: 1.3396 - val_precision: 0.0902 - val_recall: 0.3730\nEpoch 12/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m259s\x1b[0m 11ms/step - accuracy: 0.6817 - auc: 0.5649 - loss: 130041.0547 - precision: 0.1357 - recall: 0.3819 - val_accuracy: 0.8128 - val_auc: 0.5490 - val_loss: 0.7586 - val_precision: 0.2294 - val_recall: 0.3295\nEpoch 13/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m258s\x1b[0m 11ms/step - accuracy: 0.6262 - auc: 0.5517 - loss: 9647.6748 - precision: 0.1249 - recall: 0.4341 - val_accuracy: 0.5364 - val_auc: 0.6561 - val_loss: 4076.4690 - val_precision: 0.1495 - val_recall: 0.7254\nModel training complete.\n"""

revised_log_text = """LSTM model architecture redefined and compiled successfully with Adam optimizer.\nEpoch 1/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m263s\x1b[0m 11ms/step - accuracy: 0.6748 - auc_1: 0.5806 - loss: 0.7281 - precision_1: 0.1386 - recall_1: 0.4058 - val_accuracy: 0.8947 - val_auc_1: 0.5295 - val_loss: 0.4642 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nEpoch 2/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.7014 - auc_1: 0.6176 - loss: 0.6893 - precision_1: 0.1617 - recall_1: 0.4304 - val_accuracy: 0.8947 - val_auc_1: 0.6042 - val_loss: 0.5228 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nEpoch 3/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.7724 - auc_1: 0.5501 - loss: 0.7733 - precision_1: 0.1530 - recall_1: 0.2618 - val_accuracy: 0.8947 - val_auc_1: 0.5000 - val_loss: 0.3633 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nEpoch 4/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.8978 - auc_1: 0.4986 - loss: 0.9288 - precision_1: 0.0000e+00 - recall_1: 0.0000e+00 - val_accuracy: 0.8947 - val_auc_1: 0.5000 - val_loss: 0.3673 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nEpoch 5/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.8934 - auc_1: 0.4924 - loss: 0.9396 - precision_1: 0.0000e+00 - recall_1: 0.0000e+00 - val_accuracy: 0.8947 - val_auc_1: 0.5000 - val_loss: 0.3728 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nEpoch 6/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m257s\x1b[0m 11ms/step - accuracy: 0.8977 - auc_1: 0.4975 - loss: 0.8930 - precision_1: 0.0000e+00 - recall_1: 0.0000e+00 - val_accuracy: 0.8947 - val_auc_1: 0.5000 - val_loss: 0.3761 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nEpoch 7/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m256s\x1b[0m 11ms/step - accuracy: 0.8952 - auc_1: 0.4918 - loss: 0.8999 - precision_1: 0.0000e+00 - recall_1: 0.0000e+00 - val_accuracy: 0.8947 - val_auc_1: 0.3915 - val_loss: 0.3800 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nEpoch 8/20\n\x1b[1m23012/23012\x1b[0m \x1b[32m━━━━━━━━━━━━━━━━━━━━\x1b[0m\x1b[37m\u001b[0m \x1b[1m256s\x1b[0m 11ms/step - accuracy: 0.8966 - auc_1: 0.4944 - loss: 0.8862 - precision_1: 0.0000e+00 - recall_1: 0.0000e+00 - val_accuracy: 0.8947 - val_auc_1: 0.5000 - val_loss: 0.3755 - val_precision_1: 0.0000e+00 - val_recall_1: 0.0000e+00\nRevised model training complete.\n"""

# Parse the logs
initial_metrics_df = parse_training_logs(initial_log_text, model_type="initial")
revised_metrics_df = parse_training_logs(revised_log_text, model_type="revised")

# Combine the metrics into a single DataFrame
all_metrics_df = pd.concat([initial_metrics_df, revised_metrics_df], ignore_index=True)

print("Parsed metrics from both initial and revised model training:")
display(all_metrics_df.head(15)) # Display first few rows to verify
print(f"Total entries in metrics DataFrame: {len(all_metrics_df)}")
```
